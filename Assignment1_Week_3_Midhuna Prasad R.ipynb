```python
plt.subplots(figsize=(7,6), dpi=100)
sns.distplot( df.loc[df.Classification=='Iris-setosa', "SL"] , color="dodgerblue", label="Iris-setosa")
sns.distplot( df.loc[df.Classification=='Iris-virginica', "SL"] , color="orange", label="Iris-virginica")
sns.distplot( df.loc[df.Classification=='Iris-versicolor', "SL"] , color="deeppink", label="Iris-versicolor")
plt.title('PLot 1:Classified by SL')
plt.legend();
```

<h1>Data Insights:</h1>

Plot 1 shows that there is a significant amount of overlap between the classifications on SL, so it is not an effective feature for differentiating the Classifications.

```python
plt.subplots(figsize=(7,6), dpi=100)
sns.distplot( df.loc[df.Classification=='Iris-setosa', "SW"] , color="dodgerblue", label="Iris-setosa")
sns.distplot( df.loc[df.Classification=='Iris-virginica', "SW"] , color="orange", label="Iris-virginica")
sns.distplot( df.loc[df.Classification=='Iris-versicolor', "SW"] , color="deeppink", label="Iris-versicolor")
plt.title('PLot 2:Classified by SW')
plt.legend();
```

<h1>Data Insights:</h1>

Plot 2 shows that there is even higher overlap between the classifications on SW, so it is not an effective feature for differentiating the Classifications.

```python
plt.subplots(figsize=(7,6), dpi=100)
sns.distplot( df.loc[df.Classification=='Iris-setosa', "PL"] , color="dodgerblue", label="Iris-setosa")
sns.distplot( df.loc[df.Classification=='Iris-virginica', "PL"] , color="orange", label="Iris-virginica")
sns.distplot( df.loc[df.Classification=='Iris-versicolor', "PL"] , color="deeppink", label="Iris-versicolor")
plt.title('PLot 3:Classified by PL')
plt.legend();
```

<h1>Data Insights:</h1>

Plot 3 shows that PL is a good feature for differentiating the classification as it clearly separates the Classifications. The overlap is extremely less between Iris-versicolor and Iris-virginica, Iris-setosa is well separated from the rest two.

```python
plt.subplots(figsize=(7,6), dpi=100)
sns.distplot( df.loc[df.Classification=='Iris-setosa', "PW"] , color="dodgerblue", label="Iris-setosa")
sns.distplot( df.loc[df.Classification=='Iris-virginica', "PW"] , color="orange", label="Iris-virginica")
sns.distplot( df.loc[df.Classification=='Iris-versicolor', "PW"] , color="deeppink", label="Iris-versicolor")
plt.title('PLot 4:Classified by PW')
plt.legend();
```

<h1>Data Insights:</h1>

Just like Plot 3, Plot 4 also shows that PW is a good feature  for differentiating the Classifications. The overlap is significantly less between Iris-versicolor and Iris-virginica, Iris-setosa is well separated from the rest two.
